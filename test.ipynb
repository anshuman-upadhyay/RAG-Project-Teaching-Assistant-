```python
import whisper
import json
model = whisper.load_model("large-v2")
result = model.transcribe(audio = "ex-audio/8_Forms-and-input-tags-in-HTML-.mp3",
                          language = "hi",
                          task = "translate")
# print(result["text"])
with open("output.json","w") as f:
    json.dump(result,f)
    
```
